```python
import datetime
new_date = datetime.datetime(2024, 9, 6)
new_date
```

```python
RFMApproach = df.groupby('Supplier_Name').agg({'Date': lambda x: (new_date-x.max()).days,
                                            'PO_ID': lambda x: x.count(),
                                            'Value_INR': lambda x: x.sum()})

RFMApproach.rename(columns = {'Date':'Recency', 'PO_ID':'Frequency', 
                              'Value_INR':'Monetory'}, inplace=True)

RFMApproach.reset_index()
```

```python
RFMApproach.describe()
```

```python
RFMApproach.quantile(0.75)
```

```python
# Using Quantile Method to make cluster basis RFM approach
quantiles = RFMApproach.quantile(q=[0.25, 0.50, 0.75])
quantiles = quantiles.to_dict()
quantiles
```

```python
def Recency_value(x, p, d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <=d[p][0.75]:
        return 3
    else:
        return 4
```

```python
def Freq_Monetory_value(x, p, d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <=d[p][0.75]:
        return 2
    else:
        return 1
```

```python
RFMApproach.columns
```

```python
RFMApproach['R'] = RFMApproach['Recency'].apply(Recency_value, args=('Recency',quantiles))
RFMApproach['F'] = RFMApproach['Frequency'].apply(Freq_Monetory_value, args=('Frequency',quantiles))
RFMApproach['M'] = RFMApproach['Monetory'].apply(Freq_Monetory_value, args=('Monetory',quantiles))
```

```python
RFMApproach.reset_index()
```

```python
RFMApproach['RFM'] = RFMApproach[['R','F','M']].sum(axis=1)
RFMApproach.reset_index()
```

```python
RFMApproach['RFM'].nunique()
```

```python
Spend_Level = ['Top_Spend1','Top_Spend2','Top_Spend3','Top_Spend4']
score_cuts = pd.qcut(RFMApproach.RFM, q=4, labels = Spend_Level)
RFMApproach['Spend_Level'] = score_cuts.values
RFMApproach = RFMApproach.reset_index()
RFMApproach
```

```python
RFMApproach.to_csv('Final_Segmentation.csv')
```

```python
RFMApproach.columns
```

```python
final_report = RFMApproach.groupby('Spend_Level')[['Recency', 'Frequency', 'Monetory','Supplier_Name']].agg({
    'Recency':'sum', 'Frequency':'sum', 'Monetory':'sum','Supplier_Name':'nunique'}).reset_index()
final_report
```

```python
final_rep = final_report.rename(columns = {'Supplier_Name':'Supplier_count'})
final_rep['Percent'] = round(final_rep['Supplier_count']/final_rep.Supplier_count.sum()*100,2)
final_rep
```

```python
final_rep = final_rep[['Spend_Level','Supplier_count','Percent']]
final_rep
```

```python
import squarify
```

```python
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(12,8)
colors_dict = {'Top_Spend1':'red', 'Top_Spend2':'orange','Top_Spend3':'yellow', 'Top_Spend4':'green'}
squarify.plot(sizes = final_rep['Supplier_count'], color = colors_dict.values(),
              label = ['{} \n{:0.0f} supplier \n{}%'.format(*final_rep.iloc[i]) 
                       for i in range(0, len(final_rep))], alpha=0.6)

plt.title("Supplier Basis Spend Level", fontsize=25, fontweight ='bold')
plt.axis('off')
plt.show()                       
```
